```python
#Creating DataFrame separately for numerical feature and catagorical feature

numeric_feature= [feature for feature in df.columns if df[feature].dtypes!='O']
catgorical_feature= [feature for feature in df.columns if df[feature].dtypes=='O']

print('we have {} numeric_features: {}'.format(len(numeric_feature), numeric_feature))
print('we have {} catgorical_feature: {}'.format(len(catgorical_feature), catgorical_feature))
```

```python
for feature in catgorical_df:
    if feature!='Surname':
        data=df.copy()
        ax=data.groupby(feature)['Surname'].count().plot.bar()
        for container in ax.containers:
            ax.bar_label(container)
        plt.xlabel(feature)
        plt.ylabel('Surname')
        plt.title(feature)
        plt.show()
```

```python
def group_by(feature):
    df1=catgorical_df.groupby(feature,as_index=False)['Surname'].count().sort_values(by='Surname',ascending=False)
    return df1
```

```python
def plot_pie_chart(data, label_column, value_column):
    # Group the data by the label column and calculate the sum of values
    grouped_data = data.groupby(label_column)[value_column].sum()

    # Create a pie chart using pandas
    ax = grouped_data.plot(kind='pie', autopct='%1.1f%%', startangle=90)

    # Add labels to the pie chart
    ax.set_ylabel('')
    ax.set_title('Pie Chart')

 
    # Display the plot
    plt.show()
```

```python
# Churn distribution
plt.figure(figsize=(5, 5))
churn_counts = df['Exited'].value_counts()
plt.pie(churn_counts, labels=['Retained', 'Churned'], autopct='%1.1f%%', startangle=45)
plt.axis('equal')
plt.title('Churn Distribution')
plt.show()
```

```python
def churn_distribution(feature):
    plt.figure(figsize=(8, 6))
    sns.countplot(x=feature, hue='Exited', data=df)
    plt.title(f'{feature} churn ')
    plt.show()
```
